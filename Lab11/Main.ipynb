```python
import pandas as pd

def scatter(x, y, xlabel="please label me!", ylabel="please label me!"):
    df = pd.DataFrame({"x":x, "y":y})
    ax = df.plot.scatter(x="x", y="y", color="black", fontsize=16, xlim=0, ylim=0)
    ax.set_xlabel(xlabel, fontsize=16)
    ax.set_ylabel(ylabel, fontsize=16)
    ax.get_xaxis().get_major_formatter().set_scientific(False)
    ax.get_yaxis().get_major_formatter().set_scientific(False)
    
scatter([0.0, 0.0, 0.5, 1.0, 1.0, 1.0, 2.0, 2.0, 2.0, 3.0, 3.0, 3.0, 8.0, 8.0, 8.0],
        [4.1, 4.3, 4.2, 5.8, 6.0, 6.4, 8.4, 8.0, 7.9, 8.3, 8.1, 8.2, 1.9, 2.1, 1.0])
```

```python
def fact(n):
    if n==1:
        return 1
    return n * fact(n-1)

fact(4)
```

```python
def normal_fact(n):
    num = 1
    for i in range(1, n+1):
        num *= i
    return num

normal_fact(5)
```

```python
def reverse(s):
    if type(s) == str:
        if len(s) <= 1:
            return s
        return reverse(s[1:])+ (s[0])
    if type(s) == list:
        if len(s)<= 1:
            return s
        return [s[-1]]+ reverse(s[:-1])
        
reverse('covid')
```

```python
#dictionary printer
def dprint(d, indent=0):
    print("Dictionary:")
    for k in d:
        v = d[k]
        print(" " * indent, end="")
        print(k + " => ", end="")
        if type(v) == dict:
            dprint(v, indent+1)
        else:
            print(v)
            
dprint({"A": 1, "B": {"C": 2, "D": 3, "E": {"F": 4}}, "G": 5})
```

```python
rows = [
     ("alice", 5),
     ("bob", 6),
     ("alice", 7),
     ("bob", 8),
]

scores = dict()
for row in rows:
    if not row[0] in scores:
        scores[row[0]] = [row[1]]
    else:
        scores[row[0]].append(row[1])

scores
```

```python
from collections import defaultdict

rows = [
     ("alice", 5),
     ("bob", 6),
     ("alice", 7),
     ("bob", 8),
]

scores = defaultdict(list)
for row in rows:
    scores[row[0]].append(row[1]) #will create a new list if necessary

dict(scores)
```
