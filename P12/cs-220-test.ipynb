```python
def download(filename, url):
    #we don't download again if the file already exists
    if os.path.exists(filename):
        return (str(filename) + " already exists!") 
    r = requests.get(url)
    r.raise_for_status()
    f = open(filename, 'wb')
    f.write(r.content)
    f.close()      
    return (str(filename) + " created!")

download('countries.json', 'https://raw.githubusercontent.com/msyamkumar/cs220-s21-projects/master/p12/countries.json')
```

```python
#q11

#code from https://www.geeksforgeeks.org/haversine-formula-to-find-distance-between-two-points-on-a-sphere/
def haversine(lat1, lon1, lat2, lon2):  
    #distance between latitudes and longitudes in radians
    dLat = (lat2 - lat1) * math.pi / 180.0
    dLon = (lon2 - lon1) * math.pi / 180.0
    #convert latitudes to radians
    lat1 = (lat1) * math.pi / 180.0
    lat2 = (lat2) * math.pi / 180.0
    #apply formula
    a = (pow(math.sin(dLat / 2), 2) + pow(math.sin(dLon / 2), 2) * math.cos(lat1) * math.cos(lat2))
    c = 2 * math.asin(math.sqrt(a))
    return round(3956 * c, 4) #in miles 

haversine(43.070231, -89.411893, 43.074645, -89.384113)
```

```python
#q14
dicthold={}
aust=countries[countries['continent'] == 'Australia']

for i in range(len(aust)):
    countr=aust.iloc[i][4]
    dicthold[countr]={}
    for j in range(len(aust)):
        countrr=aust.iloc[j][4]
        if countr!=countrr:
            dicthold[countr][countrr]= haversine(capitals[capitals['country']==countr].iloc[0][2], capitals[capitals['country']==countr].iloc[0][3], capitals[capitals['country']==countrr].iloc[0][2], capitals[capitals['country']==countrr].iloc[0][3])
        if countr==countrr:
            dicthold[countr][countrr]= math.nan
            
df=DataFrame(dicthold)
df
```

```python
#q20
def process_json(filename):
    exampleFile = open(filename, encoding='utf-8')
    data=json.load(exampleFile)
    exampleFile.close()
    return data

download('capitals.html', 'https://raw.githubusercontent.com/msyamkumar/cs220-s21-projects/master/p12/techslides-snapshot.html')
url='https://raw.githubusercontent.com/msyamkumar/cs220-s21-projects/master/p12/techslides-snapshot.html'
html_content = requests.get(url).text
bs_obj = BeautifulSoup(html_content, "html.parser")

header=['country', 'capital', 'latitude','longitude']
souplist=[]
for i in bs_obj.find_all('tr')[1:]:
    soupdic={}
    countrydata=i.find_all('td')
    for jsn in process_json('capitals.json'):
        if jsn['country']==(countrydata[0].text):
            for j in range(len(header)):
                if header[j]=='latitude' or header[j]=='longitude':
                    soupdic[header[j]]=float(countrydata[j].text)
                    continue
                soupdic[header[j]]=countrydata[j].text
            souplist.append(soupdic)  

my_capitals = open('my_capitals.json', "w", encoding='utf-8')
json.dump(souplist,my_capitals)
my_capitals.close()

f = open('my_capitals.json', encoding='utf-8')
my_capitals_data = json.load(f)
f.close()

my_capitals_data
```
